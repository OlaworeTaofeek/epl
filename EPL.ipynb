```python
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read in the data
df = pd.read_csv('C:/Users/User/Desktop/Project/EPL 2020_2021.csv')
```

```python
# Display the first five rows by default
df.head()
```

```python
df.info()
```

```python
# Summary statistics about the dataset 
df.describe()
```

```python
# Total number of null values
df.isna().sum()
```

```python
# Create two new columns
df['MinsPerMatch'] = (df['Mins'] / df['Matches'].astype(int))
df['GoalsPerMatch'] = (df['Goals'] / df['Matches'].astype(float))
df.head()
```

```python
# Change MinPerMatch to int32
df['MinsPerMatch'] = df['MinsPerMatch'].astype(int)
```

```python
df.info()
```

```python
# Total Goals
Total_Goals = df['Goals'].sum()
Total_Goals
```

```python
# Penalty Goals
Total_PenaltyGoals = df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
# Penalty Attempts
Total_PenaltyAttempts = df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13, 6))
pl_not_scored = df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed', 'Penalties Scored']
colors = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```

```python
# Unique positions
df['Position'].unique()
```

```python
# Total FW players
df[df['Position'] == 'FW']
```

```python
# Players from diffrent nations
np.size((df['Nationality'].unique()))
```

```python
# Most players  comes from which countries
nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette('magma'))
```

```python
# Club with maximum players in their squad
df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color =sns.color_palette('viridis'))
```

```python
# Club with least players in their squad
df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
# Players based on age group
Under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20)  & (df['Age'] <= 25)]
age25_30 = df[(df['Age'] > 25)  & (df['Age'] <= 30)]
Above30 = df[df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ['<=20','>20 & <=25','>25 & <=30','>30']
plt.title('Total Players with Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
# Total under 20 players in each club
players_under20 = df[df['Age'] < 20]
players_under20['Club'].value_counts().plot(kind = 'bar',color = sns.color_palette('cubehelix'))
```

```python
# Under 20 players in Manu
players_under20[players_under20['Club'] == 'Manchester United']
```

```python
# Players under 20 in chelsea
players_under20[players_under20['Club'] == 'Chelsea']
```

```python
# Average age of players in each club
plt.figure(figsize = (12,6))
sns.boxplot(x = 'Club',y = 'Age',data = df)
plt.xticks(rotation = 90)
```

```python
# Club with average age of players
num_player = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending = False)
```

```python
# Total assists from each club
Assist_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assist_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=20)  
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 8)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
# Top 10 Assistsst(10).sort_values(ascending = False)
top_10_assists = df[['Name','Club','Assists']].nlargest(n = 10, columns = 'Assists')
top_10_assists
```

```python
# Total goals from each club
Goals_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals', fontsize=20)  
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 8)
plt.title('Plot of Clubs vs Total Goals', fontsize=20)
```

```python
# Most goals by players
top_10_goalscorers = df[['Name','Club','Goals','Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goalscorers
```

```python
# Goals per match
top_10_goals_per_match = df[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n = 10, columns = 'GoalsPerMatch')
top_10_goals_per_match
```

```python
# Pie chart - Goals with assists and without assists
plt.figure(figsize = (14,7))
assists = df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists','Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data,labels = labels,colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards
epl_yellow = df.sort_values(by = 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize = (20,6))
plt.title('Players with most yellow cards')
c = sns.barplot(x = epl_yellow['Name'],y = epl_yellow['Yellow_Cards'],label = 'Players',color = 'yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation = 45)
c
```
